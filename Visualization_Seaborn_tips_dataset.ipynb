```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.get_dataset_names()       # Data set available in package
```

```python
tips
```

#Univariante --> "one variable" data visualization

```python
# Strip plot

sns.stripplot(y = 'tip', data=tips, jitter= True)
plt.ylabel('tip($)')
plt.show()
```

# Groupinf with stripplot()


```python
sns.stripplot(x='day', y='tip', data=tips)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.stripplot(x='day', y='tip', data=tips, size=4, jitter=False)
plt.ylabel('tip($)')
plt.show()
```

```python
# Swarn plot

sns.swarmplot(x='day', y='tip', data=tips)
plt.ylabel('tips($)')
plt.show()
```

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='sex')   # hue means color
plt.ylabel('tip ($)')
plt.show()
```

```python
# Change Orientation
sns.swarmplot(x='tip', y='day', data=tips, hue='sex', orient='h')
plt.xlabel('tip ($)')
plt.show()
```

# Box and Violin plot

```python
#plt.subplot(1,2,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
```

```python
#plt.subplot(1,2,1)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
plt.subplot(1,2,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
## Combining plots

sns.violinplot(x='day', y='tip', data=tips, inner=None, color='lightgray')
sns.stripplot(x='day', y='tip', data=tips, size=4, jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

# Bivariate --> "two variables" data visualization

```python
#Joint Plot

sns.jointplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
# Density plot
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
plt.show()
```

```python
# Pair plot

sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips,hue='sex')
plt.show()
```
