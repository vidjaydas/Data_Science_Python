```python
from autoscraper import AutoScraper

url = "https://www.amazon.in/s?k=iphone&ref=nb_sb_noss_2"

wanted_list = ['New Apple iPhone 11 (64GB) - Black', "₹49,900"]

scraper = AutoScraper()
result = scraper.build(url, wanted_list)
print(result)
```

```python
scraper.get_result_similar(url, grouped=True)
```

```python
scraper.set_rule_aliases({"rule_359f" : "Price", "rule_50bn" : "Title" })
scraper.keep_rules(["rule_50bn", "rule_359f"])
scraper.save('Amazon_search')
```
