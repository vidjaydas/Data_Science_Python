```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
#Conda install -c conda-forge mlxtend
```

```python
titanic = pd.read_csv('/content/Titanic.csv')
titanic.head()
```

#Pre-Processing

####As the data is not in transaction formation We are using transaction Encoder

```python
df = pd.get_dummies(titanic)
df.head()
```

#Apriori Algorithm

```python
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric = 'lift', min_threshold=0.7)
rules

#Ignore the leverage	conviction columns. not useful
#if we increase the threshold value we will get refined data
```

### An Leverage value of 0 indicates independence. Range will be [-1 1]

###A high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]

```python
rules.soft_values('lift', ascending = False)[:20]
```

```python
rules[rules.lift>1]
```
